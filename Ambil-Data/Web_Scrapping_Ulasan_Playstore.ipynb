```python
from google_play_scraper import app, reviews
import pandas as pd
import datetime
from google_play_scraper import Sort
```

```python
# Fungsi untuk mengambil ulasan aplikasi
def get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST, filter_score_with=None):
    try:
        # Ambil ulasan aplikasi
        result, continuation_token = reviews(
            app_id,
            lang=lang,
            count=count,
            sort=sort,
            filter_score_with=filter_score_with
        )

        # Simpan hasil ulasan dalam DataFrame
        df = pd.DataFrame(result)

        # Jika ingin menambahkan waktu ambil ulasan
        df['date_fetched'] = datetime.datetime.now()

        return df
    except Exception as e:
        print(f"Terjadi kesalahan saat mengambil data ulasan: {e}")
        return None

# ID aplikasi yang ingin diambil ulasannya
app_id = 'com.twitter.android'

# Mengambil ulasan aplikasi
reviews_df = get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST)

# Tampilkan DataFrame untuk memastikan pengambilan data berhasil
reviews_df.head()
```

```python
# Fungsi untuk mengambil ulasan aplikasi
def get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST, filter_score_with=None):
    try:
        # Ambil ulasan aplikasi
        result, continuation_token = reviews(
            app_id,
            lang=lang,
            count=count,
            sort=sort,
            filter_score_with=filter_score_with
        )

        # Simpan hasil ulasan dalam DataFrame
        df = pd.DataFrame(result)

        # Jika ingin menambahkan waktu ambil ulasan
        df['date_fetched'] = datetime.datetime.now()

        return df
    except Exception as e:
        print(f"Terjadi kesalahan saat mengambil data ulasan: {e}")
        return None

# Fungsi untuk menghindari duplikasi dengan membandingkan ID ulasan
def avoid_duplicates(new_df, old_df):
    if old_df is not None:
        # Gabungkan ulasan lama dan baru, tapi pastikan ID ulasan unik
        combined_df = pd.concat([old_df, new_df])
        combined_df.drop_duplicates(subset='reviewId', keep='last', inplace=True)
        return combined_df
    else:
        return new_df

# ID aplikasi yang ingin diambil ulasannya
app_id = 'com.twitter.android'

# Mengambil ulasan aplikasi
reviews_df_new = get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST)

# Misalnya, ini DataFrame lama yang telah diambil sebelumnya
# reviews_df_old = pd.read_csv('twitter_reviews.csv')  # Jika ada file CSV lama

# Hindari duplikasi jika data sebelumnya ada
reviews_df = avoid_duplicates(reviews_df_new, None)  # Ganti 'None' dengan 'reviews_df_old' jika ada data sebelumnya

# Tampilkan DataFrame untuk memastikan tidak ada duplikasi
print(reviews_df.head())
```

```python
# Fungsi untuk mengambil ulasan aplikasi
def get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST, filter_score_with=None):
    try:
        # Ambil ulasan aplikasi
        result, continuation_token = reviews(
            app_id,
            lang=lang,
            count=count,
            sort=sort,
            filter_score_with=filter_score_with
        )

        # Simpan hasil ulasan dalam DataFrame
        df = pd.DataFrame(result)

        # Jika ingin menambahkan waktu ambil ulasan
        df['date_fetched'] = datetime.datetime.now()

        return df
    except Exception as e:
        print(f"Terjadi kesalahan saat mengambil data ulasan: {e}")
        return None

# ID aplikasi yang ingin diambil ulasannya
app_id = 'com.facebook.katana'

# Mengambil ulasan aplikasi
reviews_df = get_reviews(app_id, lang='id', count=2000, sort=Sort.NEWEST)

# Tampilkan DataFrame untuk memastikan pengambilan data berhasil
reviews_df.head()
```

```python
# Fungsi untuk mengupload DataFrame ke CSV
def upload_to_csv(df, file_name='facebook_reviews.csv'):
    try:
        if df is not None:
            # Menyimpan DataFrame dalam file CSV
            df.to_csv(file_name, index=False, encoding='utf-8')
            print(f"Data berhasil disimpan ke {file_name}")
        else:
            print("Data tidak tersedia, tidak dapat disimpan ke CSV.")
    except Exception as e:
        print(f"Terjadi kesalahan saat menyimpan data ke CSV: {e}")

# Mengupload DataFrame ke CSV jika berhasil mengambil data
upload_to_csv(reviews_df, 'facebook_reviews.csv')
```
